```python
import glob
import pandas as pd
import seaborn as sns 

sns.set_style("white")
```

```python
'./data/summarized/ADEM/RCP8.5/MEAN SLR RCP8.5.CSV'.split('/')[5]
```

```python
for file in glob.glob('./data/summarized/**/**/*.CSV'):
    df = pd.read_csv(file)

    cols = ['Date', 'Developed Dry Land', 'Undeveloped Dry Land', 'Swamp',
           'Cypress Swamp', 'Inland-Fresh Marsh', 'Tidal-Fresh Marsh',
           'Trans. Salt Marsh', 'Regularly-Flooded Marsh', 'Mangrove',
           'Estuarine Beach', 'Tidal Flat', 'Ocean Beach', 'Ocean Flat',
           'Rocky Intertidal', 'Inland Open Water', 'Riverine Tidal',
           'Estuarine Open Water', 'Tidal Creek', 'Open Ocean',
           'Irreg.-Flooded Marsh', 'Inland Shore', 'Tidal Swamp']
    categories = {
        'MANGROVE'   : ["Trans. Salt Marsh", "Mangrove", "Irreg.-Flooded Marsh"],
        'TIDAL_FLAT' : ["Regularly-Flooded Marsh", "Estuarine Beach", "Tidal Flat", "Tidal Swamp"],
        'REEF'       : ["Ocean Flat", "Rocky Intertidal"]
    }
    df = df[cols]
    df = df.loc[:, (df != 0).any(axis=0)]
    # df.head()
    
    # remove '2011' row
    df = df.drop(labels=1 , axis=0)
    # rename first row (Date = 0) to 2011
    df.at[0,'Date'] = 2011
    
    for category, habitats in categories.items():
        df[category] = df[habitats].sum(axis=1)
        df = df.drop(labels=habitats, axis=1)
        
    
    # plot absolute change
    _df = df.iloc[::-1]
    ax = _df.drop(['Date', 'Open Ocean'], axis=1).T.plot(kind='barh', 
                   width=0.7, 
                   figsize=(10,10), 
                   color=[(0.3,0.4,0.7), # bottom
                          (0.3,0.5,0.7),
                          (0.3,0.58,0.8),
                          (0.3,0.58,0.85),
                          (0.3,0.58,0.95), # top
                          ])
        
    fileparts = file.split('/')
    type_of_dem = fileparts[3]
    scenario = fileparts[5]
    title = f'Change in Habitat ({type_of_dem}, {scenario})'
    ax.set_title(title)
    ax.legend(_df['Date'])
    ax.grid(True, axis='x')

    fig = ax.get_figure()
    fig.savefig(f'charts/{title}.pdf', bbox_inches='tight')
    
    # plot percentage change
    df_pct_change_against_2011 = df.apply(lambda x: x.div(x.iloc[0]).subtract(1).mul(100))
    _df = df_pct_change_against_2011.iloc[::-1]

    ax = _df.drop(['Date', 'Open Ocean'], axis=1).T.plot(kind='barh', 
                   width=0.7, 
                   figsize=(10,10), 
                   color=[(0.3,0.4,0.7), # bottom
                          (0.3,0.5,0.7),
                          (0.3,0.58,0.8),
                          (0.3,0.58,0.85),
                          (0.3,0.58,0.95), # top
                          ])

    title = f'Change in Habitat ({type_of_dem}, {scenario}) (in percentage) (base year = 2011)'
    ax.set_title(title)
    ax.legend(df['Date'].iloc[::-1])
    ax.grid(True, axis='x')
    fig = ax.get_figure()

    fig.savefig(f'charts/{title}.pdf', bbox_inches='tight')
```
